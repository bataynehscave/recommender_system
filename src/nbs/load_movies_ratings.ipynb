```python
import math
```

```python
df['value'] = df['rating'].apply(lambda x: math.ceil(x))
```

```python
df.head()
```

```python
df['user_id'] = df['userId']
```

```python
df['object_id'] = df['movieId']
```

```python
cols = ['user_id', 'value', 'object_id']
transformed_df = df[cols].copy()
```

```python
transformed_df.head()
```

```python
rating_records = transformed_df.to_dict('records')
```

```python
from ratings.models import Rating
qs = Rating.objects.all()
qs.delete()
```

```python
qs
```

```python
from django.contrib.contenttypes.models import ContentType
ctype = ContentType.objects.get(app_label='movies', model = 'movie')
```

```python
ctype
```

```python
new_rating = []
for r in rating_records:
    r['content_type'] = ctype
    new_rating.append(Rating(**r))
```

```python
Rating.objects.bulk_create(new_rating, batch_size = 1000)
```
